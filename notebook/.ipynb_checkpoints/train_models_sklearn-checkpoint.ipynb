### `Données`

```python
# Importer les données 
# cancer = pd.read_csv("data/process/cancer_clean.csv", sep=",")
cancer = pd.read_csv("cancer_clean.csv", sep=",")
```

### `Split data`

```python
from sklearn.model_selection import train_test_split

# X et y 
X = cancer.drop("diagnosis", axis=1)
y = cancer["diagnosis"]

# X_train, X_test, y_train, y_test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=3)
```

---

### `Calculer la variance pour chaque feature`

```python
from sklearn.preprocessing import MinMaxScaler

# Minmaxscaler 
MinMax_scaler = MinMaxScaler()

# Mise à l'éechelle des variables 
X_scaler = MinMax_scaler.fit_transform(X)
```

```python
fig, ax = plt.subplots()

x = X.columns
y = X_scaler.var(axis=0) 

ax.bar(x, y, width=0.2)
ax.set_xlabel("Features")
ax.set_ylabel("variance")
ax.set_ylim(0, 0.1)

for index, value in enumerate(y): 
    plt.text(x=index, y=value+0.001, 
            s=str(round(value, 3)), ha="center")

fig.autofmt_xdate()
plt.tight_layout()
```

---

### `Standardisation` 

```python
# Minmaxscaler 
MinMax_scaler = MinMaxScaler()

# Mise à l'échelle des variables 
X_train_scaler = MinMax_scaler.fit_transform(X_train)
X_test_scaler = MinMax_scaler.transform(X_test)

# as dataframe
X_train_scaler = pd.DataFrame(X_train_scaler,columns=X.columns) 
X_test_scaler = pd.DataFrame(X_test_scaler,columns=X.columns) 

# Supprimer la variable "fractal_dimension_mean"
# X_train_scaler = X_train_scaler1.drop(["fractal_dimension_mean"], axis=1)
# X_test_scaler = X_test_scaler1.drop(["fractal_dimension_mean"], axis=1)
```

```python
# shpae X_train_scaler1 & X_test_scaler1
print(X_train_scaler.shape), print(X_test_scaler.shape)
```

```python
X_train_scaler.ndim , X_test_scaler.ndim
```

```python
X_train_scaler = X_train_scaler.drop("symmetry_worst", axis=1)
X_test_scaler = X_test_scaler.drop("symmetry_worst", axis=1)
```

```python
X_train_scaler
```

### `Entrainement des algorithmes de classification`

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.model_selection import cross_val_score, validation_curve, StratifiedKFold
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier, GradientBoostingClassifier
```

--- 

### `Régression logistique`

```python
# Avec toutes les variables 
logistic_all = LogisticRegression(random_state=179)
logistic_all = logistic_all.fit(X_train_scaler, y_train)
y_pred_all = logistic_all.predict(X_test_scaler)

print(classification_report(y_test, y_pred_all))
```

---

### `DecisionTreeClassifier`

```python
# Avec toutes les variables 
tree = DecisionTreeClassifier(random_state=179)
tree = tree.fit(X_train_scaler, y_train)
accuracy_all = accuracy_score(y_test, tree.predict(X_test_scaler))
accuracy_all

print(classification_report(y_test, tree.predict(X_test_scaler)))
```

---

### `RandomForestClassifier`

```python
# Avec toutes les variables 
forest = RandomForestClassifier(random_state=179)
forest = forest.fit(X_train_scaler, y_train)
accuracy_all = accuracy_score(y_test, forest.predict(X_test_scaler))
accuracy_all

print(classification_report(y_test, forest.predict(X_test_scaler)))
```

---

### `ExtraTreesClassifier`

```python
# Avec toutes les données 
extra_tree_all = ExtraTreesClassifier(random_state=179)
extra_tree_all = extra_tree_all.fit(X_train_scaler, y_train)
y_pred_all_extra = extra_tree_all.predict(X_test_scaler)


# Accuracy 
accuracy_all_extra = accuracy_score(y_test, y_pred_all_extra)

print(classification_report(y_test, y_pred_all_extra))
```

---

### `GradientBoostingClassifier`

```python
# Avec toutes les données 
booster = GradientBoostingClassifier(random_state=179)
booster = booster.fit(X_train_scaler, y_train)
accuracy_all = accuracy_score(y_test, booster.predict(X_test_scaler))
accuracy_all

print(classification_report(y_test, booster.predict(X_test_scaler)))
```

---

### `AdaBoostClassifier`

```python
# Avec toutes les données 
ada = AdaBoostClassifier(random_state=179)
ada = ada.fit(X_train_scaler, y_train)
accuracy_all = accuracy_score(y_test, ada.predict(X_test_scaler))
accuracy_all

print(classification_report(y_test, ada.predict(X_test_scaler)))
```

### `KNeighborsClassifier`

```python
from sklearn.neighbors import KNeighborsClassifier

# Avec toutes les variables 
knn = KNeighborsClassifier(n_neighbors=30)
knn.fit(X_train_scaler, y_train)
y_pred_knn = knn.predict(X_test_scaler)

# Accuracy
accuracy_knn = accuracy_score(y_test, y_pred_knn)

print(classification_report(y_test, y_pred_knn))
```

### `Sélection de modèles`

À l’issue de la phase d’entraînement, les modèles retenus sont `LogisticRegression` et `KNeighborsClassifier`.

**Justification** :

L’objectif principal de cette étude est la minimisation des faux négatifs, ce qui implique de privilégier des modèles présentant un rappel (recall) supérieur à 90 %, critère critique dans un contexte de diagnostic médical. Le modèle sélectionné offre un compromis satisfaisant entre rappel, spécificité et précision, garantissant à la fois une bonne capacité de détection des cas malins et une limitation des faux positifs. 

- **Important** : Nous n'avons pas éffectué une sélection de variables.
  

---

### `Learning curve (accuracy) - Régression logistique`

```python
from sklearn.model_selection import learning_curve

# Modèle
logistic_all = LogisticRegression(random_state=179, max_iter=500)

# Calcul des learning curves
train_sizes, train_scores, test_scores = learning_curve(logistic_all, X_train_scaler, y_train,
    cv=5, scoring='accuracy', train_sizes=np.linspace(0.1, 1.0, 10), n_jobs=-1)

# Moyennes et écarts-types
train_mean = np.mean(train_scores, axis=1)
train_std  = np.std(train_scores, axis=1)
test_mean  = np.mean(test_scores, axis=1)
test_std   = np.std(test_scores, axis=1)

# Plot
plt.figure(figsize=(10,6))
plt.plot(train_sizes, train_mean, 'o-', color='blue', label='Score entraînement')
plt.plot(train_sizes, test_mean, 'o-', color='orange', label='Score validation')

plt.fill_between(train_sizes, train_mean-train_std, train_mean+train_std, alpha=0.2, color='blue')
plt.fill_between(train_sizes, test_mean-test_std, test_mean+test_std, alpha=0.2, color='orange')

plt.title("Courbe d’apprentissage – Régression Logistique")
plt.xlabel("Taille de l’échantillon d’entraînement")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)
plt.show()
```

### `Learning curve (fonction de coût) – Régression Logistique`

```python
# Modèle
logistic_all = LogisticRegression(random_state=179, max_iter=500)

# Learning curve avec fonction de coût (log-loss)
train_sizes, train_scores, test_scores = learning_curve(logistic_all, X_train_scaler, y_train,
    cv=5, scoring='neg_log_loss', train_sizes=np.linspace(0.1, 1.0, 10), n_jobs=-1)

# On remet la vraie loss (valeurs positives)
train_loss = -np.mean(train_scores, axis=1)
train_std  = np.std(train_scores, axis=1)
test_loss  = -np.mean(test_scores, axis=1)
test_std   = np.std(test_scores, axis=1)

# Plot
plt.figure(figsize=(10,6))
plt.plot(train_sizes, train_loss, 'o-', color='blue', label='Loss entraînement')
plt.plot(train_sizes, test_loss, 'o-', color='orange', label='Loss validation')

plt.fill_between(train_sizes, train_loss-train_std, train_loss+train_std, alpha=0.2, color='blue')
plt.fill_between(train_sizes, test_loss-test_std, test_loss+test_std, alpha=0.2, color='orange')

plt.title("Courbe d’apprentissage (fonction de coût) – Régression Logistique")
plt.xlabel("Taille de l’échantillon d’entraînement")
plt.ylabel("Log-loss")
plt.legend()
plt.grid(True)
plt.show()
```
